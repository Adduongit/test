```python
from IPython.display import display, Image

# Create an input box for the user's name
name_input = widgets.Text(
    value='',
    placeholder='Enter your name',
    description='Your Name because I FWU heavy:',
    disabled=False
)

# Create a button to submit the input
submit_button = widgets.Button(
    description='Submit',
    button_style='success',  # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to submit your name'
)

# Create an empty output area
output = widgets.Output()

# Tulip image URL (you can change this with any tulip image URL)
tulip_image = Image(url="https://i.pinimg.com/736x/95/a4/46/95a446949b7724fcafab75d45964d6ca.jpg", width=300)

# Define the function that will update the greeting message
def on_button_click(b):
    with output:
        # Get the name entered by the user
        user_name = name_input.value
        if user_name:
            print(f"Happy Valentine's Day, {user_name}!")
        else:
            print("Happy Valentine's Day!")

# Link the button's click event to the function
submit_button.on_click(on_button_click)

# Display the tulip image, input field, button, and output area
display(tulip_image, name_input, submit_button, output)
```
